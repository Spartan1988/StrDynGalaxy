```python
# Conversión de unidades para l y b (grados a radianes)
import numpy as np
l = np.deg2rad(IDPos['L'])
b = np.deg2rad(IDPos['B'])
# Separando las columnas necesaciar para hacer los cálculos (R_Sun, R_gc)
Rsun = 8.0
r = IDPos['R_Sun']
```

Para o cálculo de $\cos\psi_{i}$ de cada Aglomerado usamos:

\begin{equation}
\cos\psi_{i} = \dfrac{R_{\odot}\cos b_{i}\sin l_{i}}{\sqrt{(r_{i}\cos b_{i}\cos {l_{i}}-R_{\odot}^{2})+ r_{i}^{2}\cos^{2}b_{i}\sin^{2}l_{i}}}
\end{equation}

Onde,

$b_{i}$, $l_{i}$  são as coordenadas Galácticas do $AG_{i}$ / are the standard revised galactic coordinates. <br>
$R_{\odot}$ é a distância Galactocêntrica do Sol / is the distance of the Galactic Centre from the Sun. (Usar $R_{\odot}=8kpc$) <br>
$r_{i}$ é a distância heliocêntrica de $AG_{i}$ / is the distance of the cluster from the Sun.

```python
cosb = np.cos(b)
cosl = np.cos(l)
sinb = np.sin(b)
sinl = np.sin(l)
cosPsi_num = Rsun*cosb*sinl
cosPsi_dem2 = (r*cosb*cosl-Rsun)**2 + (r*cosb*sinl)**2 
cosPsi = cosPsi_num/np.sqrt(cosPsi_dem2)
DFcosPsi = pd.DataFrame({'Cos(Psi)': cosPsi})
DFcosPsi
```

\begin{align}
V_{GSR}^{AG} &= V_{r} + 8.5\cos b\cos l + 13.38\cos b\sin l + 6.49\cos b\sin l\\
V_{GSR}^{LSR} &\approx 220
\end{align}

```python
# Cálculo de Vag_GSR
Vr = Vel['v_r']
Vag_GSR = Vr + 8.5*cosb*cosl + 13.38*cosb*sinl + 6.49*cosb*sinl
DFVag_GSR = pd.DataFrame({'Vag_GSR' : Vag_GSR})
Vlsr_GSR = 220 
DFVag_GSR
```

Calculamos os fatores de proyeção faltantes ($\cos\phi_{i}$, $\cos\lambda_{i}$):
\begin{align}
\cos\lambda_{i} &= \cos b_{i}\sin l_{i}\\
\cos\phi_{i} &= r_{i} - R_{\odot}\cos b_{i}\sin l_{i}[r_{i}^{2} + R_{\odot}^{2} + 2r_{i}R_{\odot}\cos b_{i}\cos l_{i}]^{-1/2}
\end{align}

```python
coslambda = cosb*sinl
DFcoslambda = pd.DataFrame({'Cos(lambda)' : coslambda})
cosphi = r - Rsun*cosb*sinl*(r**2 + Rsun**2 + 2*r*Rsun*cosb*cosl)**(-1/2)
DFcosphi = pd.DataFrame({'Cos(phi)' : cosphi})
```

```python
# Creating a DataFrame with the speficic data needed from the last Tables
MyTable = pd.concat([IDPos['ID'],IDPos['Name'],IDPos['R_gc'],Metallicity['[Fe/H]'],Vel['v_r'],Vel['v_LSR'],DFVag_GSR,DFcosPsi,DFcoslambda,DFcosphi], axis = 1)
MyTable['Vlsr_GSR'] = 220
MyTable.loc[0,'Vlsr_GSR'] = 'NaN'
MyTable
#MyTable = My.drop(MyTable.index[0])
```

```python
Poor = MyTable['[Fe/H]'] < -0.8
MetalPoor = MyTable[Poor]
MetalRich = MyTable[~Poor]
MetalPoor
```

Finalmente, podemos estimar a velocidade de rotação com a siguente formula:
\begin{equation}
V_{rot} = \dfrac{\Sigma\cos\psi_{i}(V_{o,i}+v_{\odot}\cos\lambda_{i})}{\Sigma\cos^{2}\psi_{i}}
\end{equation}

```python
Vrot_Num_i = cosPsi*(Vag_GSR + Vlsr_GSR*coslambda)
Vrot_Dem_i = cosPsi**2
Vrot_Num = np.sum(Vrot_Num_i) # Suma de elementos en el Numerador
Vrot_Dem = np.sum(Vrot_Dem_i) # Suma de elementos en el Denominador 
Vrot = Vrot_Num/Vrot_Dem
print "Total Vrot = ", Vrot
print "Atraso Rotacional = ", Vrot - Vlsr_GSR
```

```python
cosPsi1 = MetalPoor['Cos(Psi)']
Vag_GSR1 = MetalPoor['Vag_GSR']
Vlsr_GSR1 = MetalPoor['Vlsr_GSR']
coslambda1 = MetalPoor['Cos(lambda)']
MPVrot_Num_i = cosPsi1*(Vag_GSR1 + Vlsr_GSR1*coslambda1)
MPVrot_Dem_i = cosPsi1**2
MPVrot_Num = np.sum(MPVrot_Num_i) # Suma de elementos en el Numerador
MPVrot_Dem = np.sum(MPVrot_Dem_i) # Suma de elementos en el Denominador 
MPVrot = MPVrot_Num/MPVrot_Dem
print "Metal-poor Vrot = ", MPVrot
print "Atraso Rotacional = ", MPVrot - Vlsr_GSR
```

```python
cosPsi2 = MetalRich['Cos(Psi)']
Vag_GSR2 = MetalRich['Vag_GSR']
Vlsr_GSR2 = MetalRich['Vlsr_GSR']
coslambda2 = MetalRich['Cos(lambda)']
MRVrot_Num_i = cosPsi2*(Vag_GSR2 + Vlsr_GSR2*coslambda2)
MRVrot_Dem_i = cosPsi2**2
MRVrot_Num = np.sum(MRVrot_Num_i) # Suma de elementos en el Numerador
MRVrot_Dem = np.sum(MRVrot_Dem_i) # Suma de elementos en el Denominador 
MRVrot = MRVrot_Num/MRVrot_Dem
print "Metal-rich Vrot = ", MRVrot
print "Atraso Rotacional = ", MRVrot - Vlsr_GSR
```
